# Klasifikasi Jual Beli Rumah Berdasarkan Kategori Harga 

## 1. Pendahuluan

### Latar Belakang

Pada era kini rumah memiliki peran penting dalam masyarakat, saat seseorang ingin berinvestasi pada sebuah rumah, perlu mengetahui harga rumah sesuai kriteria yang diinginkan. Berjalannya waktu rumah berkembang menjadi bentuk investasi yang menarik, sehingga membuat daftar harga rumah yang beragam sesuai faktor utama untuk dipertimbangkan dalam harga rumah, oleh karena itu untuk dapat memprediksi rumah berdasarkan faktor yang disediakan. Melihat pentingnya rumah pada era masa kini, maka dari itu saya memiliki peluang yang lebih besar untuk membuka bisnis jual beli rumah ini. Dengan tujuan memberikan solusi yang terbaik dan memenuhi kebutuhan masyarakat. 

Visi dan Misi 

Visi

Membangun Entrepreneur Bisnis Properti & memberikan kemudahan transaksi properti di Indonesia

### Strategi Bisnis

Strategi bisnis yang dilakukan yaitu untuk memberikan prediksi tentang harga jual rumah yang digunakan untuk customer sebagai acuan dalam pembelian rumah. Setelah dilakukan prediksi harga jual beli rumah dapat memberi informasi kepada customer bahwa terdapat beberapa kategori dalam harga jual beli rumah. Oleh karena itu dapat membantu customer dalam mengambil keputusan untuk menentukan pilihan rumah sesuai anggaran mereka dan dapat mengetahui kelayakan rumah tersebut.

### Target Bisnis

Dalam penelitian ini bisnis bergerak dalam bidang properti, maka dari itu target bisnis menarik minat customer dalam pembelian rumah berdasarkan kategori. Selain itu juga menjalin kerjasama antar bisnis properti lain agar bisnis yang dilakukan berkembang lebih pesat.

### Tujuan Bisnis

Diadakan bisnis ini untuk meningkatkan pengetahuan mengenai harga jual rumah di pasaran. Sehingga menarik minat customer dalam membeli rumah yang lebih cerdas dan terinformasi. Dengan bisnis ini dapat membantu customer dalam menentukan properti sesuai anggaran dan kebutuhan yang diinginkan.

### Rumusan Masalah 

Masalah yang bisa terjadi dalam bisnis ini adalah kurangnya minat customer dalam membeli rumah yang mungkin saja disebabkan oleh faktor ketidakpastian ekonomi dan keterbatasan informasi. Selain itu bisnis properti memiliki daya saing dengan bisnis properti yang lain.
